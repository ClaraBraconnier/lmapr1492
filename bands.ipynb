```python
print("**Calcul de la masse effective**")
print("\n")
print("Pour calculer la masse effective, on approxime la courbe par une parabole (y=ax^2+bx+c)et on prend sa drive seconde afin de l'injecter dans la formume m*=hbar^2/drive seconde. L'approximation de la courbe par une parabole nous permet de considrer m* comme constante")

#Constates utiles pour la suite
A = 10**(-10)
eV = 1.602176563*10**(-19)
hbar = 1.054571818*10**(-34)
print("\n")
#Approximation de la courbe de la dernire bande de valence par une parabole. 
#On chercher d'abord le point de la bande de plus haute nergie : 
print("*Dernire bande de valence*")
y_max = max(max(branche1.get('1')[14]),max(branche2.get('1')[14]),max(branche3.get('1')[14]),max(branche4.get('1')[14]),max(branche5.get('1')[14]),max(branche6.get('1')[14]),max(branche7.get('1')[14]),max(branche8.get('1')[14]),max(branche9.get('1')[14]))
x_max = (distance)[6]
x_p = (distance)[4]
y_p = -0.071
print("Le max de la dernire bande de valence se trouve en :(",x_max,',',y_max,"). Il correspondra  prendra le sommet de la parabole (ce maximum survient 2 fois sur la bande de valende mais j'ai dcid de prendre celui se situant en L)")
print('\n')
print("On choisit un autre point pour faire l'approximation parabolique, j'ai choisi le point : (",x_p,",", y_p,")" )
print("J'ai choisi ce point grce au interactive plots de Materials Project.")
print('\n')
print("On a donc deux points de la parabole et on connat la drive premire en son sommet (=0). On a donc 3 quations  trois inconnues pour dterminer les coefficients a, b et c.")
a = (y_p-y_max)/(x_p-x_max)**2
b = -2*a*x_max
c = y_max - b*x_max - a*(x_max)**2
print("En rsolvant ces quations, on obtient la parabole d'quation : y =",a,"x^2+",b,"x",c)
def f(x):
    return a*x**2+b*x+c
x=np.arange(0.00, (distance)[14],0.01)
y = f(x)
print("\n")

#Approximation de la courbe de la premire bande de conduction par une parabole. 
#On chercher d'abord le point de la bande de plus basse nergie : 
print("*Premire bande de conduction*")
y_min = min(min(branche1.get('1')[15]),min(branche2.get('1')[15]),min(branche3.get('1')[15]),min(branche4.get('1')[15]),min(branche5.get('1')[15]),min(branche6.get('1')[15]),min(branche7.get('1')[15]),max(branche8.get('1')[15]),min(branche9.get('1')[15]))
x_min = (distance)[6]
x_pc = (distance)[4]
y_pc = 5.64
print("Le min de la premire bande de conduction se trouve en :(",x_min,',',y_min,"). Il correspondra  prendra le sommet de la parabole (ce maximum survient 2 fois sur la bande de valende mais j'ai dcid de prendre celui se situant en L)")
print('\n')
print("On choisit un autre point pour faire l'approximation parabolique, j'ai choisi le point : (",x_pc,",", y_pc,")" )
print("J'ai choisi ce point grce au interactive plots de Materials Project.")
print('\n')
print("On a donc deux points de la parabole et on connat la drive premire en son sommet (=0). On a donc 3 quations  trois inconnues pour dterminer les coefficients a, b et c.")
a2 = (y_pc-y_min)/(x_pc-x_min)**2
b2 = -2*a2*x_min
c2 = y_min - b2*x_min - a2*(x_min)**2
print("En rsolvant ces quations, on obtient la parabole d'quation : y =",a2,"x^2",b2,"x+",c2)
def g(z):
    return a2*z**2+b2*z+c2
z=np.arange(0.00,(distance)[14],0.01)
y2 = g(z)
print("\n")
print(Ces paraboles sont reprsentes en rouge sur le graphe de la structure de bande.)

#Plot le graphe et les paraboles
Graph.get_plot()
plt.plot(x,y,"r-")
plt.plot(z,y2,"r-")
plt.show()
```
