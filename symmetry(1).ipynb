```python
#Tâche 6 : [NaSmO2] étudier l’effet de 3 éléments de symétrie différents,
#chacun pour 1 atome différent

from pymatgen.io.cif import CifParser 
parser = CifParser("NaSmO2_mp-756034_symmetrized.cif") #importe le fichier CIF
from pymatgen.ext.matproj import MPRester 
import matplotlib.image as mpimg
import numpy as np
import matplotlib.pyplot as plt
from pymatgen.core.operations import SymmOp #importe la classe SymmOp qui va permettre d'effectuer les opérations de symétrie
print("**Etudier l’effet de 3 éléments de symétrie différents, chacun pour 1 atome différent**")
#Création de 3 opérations de symétrie : 
operation1=SymmOp.from_xyz_string('-x,-y,-z') #opération d'inversion
operation2=SymmOp.from_xyz_string('y,-x+y,-z') #centre d'inversion situé en (0,0,0)
operation3=SymmOp.from_xyz_string('-y,-x,z')#opération de réflexion par rapport à un plan miroir  
 
#Choix de 3 atomes différents de NaSm02 :
Na=[0.67,0.33,0.33]
Sm=[0.67,0.33,0.83]
O=[0.67,0.33,0.57]
 
#Application de ces opérations sur les différents atomes : 
new_Na=operation1.operate(Na)
new_Sm=operation2.operate(Sm)
new_O=operation3.operate(O)
 
#Affiche les résultats : 
print("\n")
print("L'atome initialement situé en", Na, "suite à l'opération",operation1)
print("se trouve en",new_Na,"il a donc subit une opération inverse")
img1 = mpimg.imread("operation1.png")
plt.imshow(img1)
plt.show()
print("L'atome initialement situé en", Sm, "suite à l'opération",operation2) 
print("se trouve en",new_Sm,"il a donc subit une opération donc le centre d'inversion est situé en (0,0,0)")
img2 = mpimg.imread("operation2.png")
plt.imshow(img2)
plt.show()
print("L'atome initialement situé en", O, "suite à l'opération",operation3)
print("se trouve en",new_O,"il a donc subit une opération de réflexion par rapport à un plan miroir")
img3 = mpimg.imread("operation3.png")
plt.imshow(img3)
plt.show()
print("REM : les images sont également dans le répertoire où elles sont de meilleure qualité")
```
