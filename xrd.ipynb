```python
#Tche 7 : [NaSmO2] dterminer les 3 premiers pics du diffractogramme correspondant (en
#indiquant les indices hkl associs) pour un rayonnement X incident Cu K
#de longueur donde =1.54060 .

from pymatgen.io.cif import CifParser 
parser = CifParser("NaSmO2_mp-756034_symmetrized.cif") #importe le fichier CIF
from pymatgen.ext.matproj import MPRester 
from pymatgen.analysis.diffraction.xrd import XRDCalculator  
from pymatgen.symmetry.analyzer import SpacegroupAnalyzer
#from pymatgen.symmetry.analyzer import Lattice
with MPRester("blOWflU3npEF7GSb") as m:
    
# Cre un objet structure avec la structure du matriau NaSmO2
    structure = m.get_structure_by_material_id("mp-756034")
#Soucis sur Material Project : cette ligne permet de faire en sorte que 
#la strtucture respecte a=b=c et alpha=beta=gamme
    new_s=SpacegroupAnalyzer(structure).get_primitive_standard_structure()
    
#Creation d'un calculateur XRDCalculator pour un rayonnement X incident Cu K, une 
#prcision de la symtrie pour le raffinement de la structure de 0.1 : 
    xrdCalc = XRDCalculator(wavelength='CuKa', symprec=0.1)
    
#Cre un objet XRDPattern avec le spectre de diffraction : 
    spctre = xrdCalc.get_pattern(new_s)
    
#Dtermine les indices hkl des 3 premiers pics du diffractogramme :
    premier_pic = (spctre.hkls)[1][0]["hkl"]
    deuxieme_pic = (spctre.hkls)[2][0]["hkl"]
    troisieme_pic = (spctre.hkls)[3][0]["hkl"]
    
#Affiche le spectre : 
    xrdCalc.show_plot(new_s)
    
#Affiche les indices hkl associs aux 3 premiers pics : 
    print("Indices hkl associ au premier pic du diffractogramme :", premier_pic)
    print("Indices hkl associ au deuxime pic du diffractogramme :", deuxieme_pic)
    print("Indices hkl associ au troisieme pic du diffractogramme", troisieme_pic)
    
```
